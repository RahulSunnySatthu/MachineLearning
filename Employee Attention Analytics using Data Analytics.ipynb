```python
# as attrition lables are in yes or no , for comparision we need numerical data to use LabelEncoder
from sklearn.preprocessing import LabelEncoder
sc=LabelEncoder()
df['transformed_Attirition']=sc.fit_transform(df['Attrition'])
df[['Attrition','transformed_Attirition']].head()
```

```python
#Bivariate Analysis
import seaborn as sn
sn.boxplot(x=df["transformed_Attirition"],
             y=df["MonthlyIncome"],
             data=df
             )
```

From the boxplot, we can see that employees with lower monthly income have higher attrition, whereas higher-paid employees tend to stay longer.
