```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
paragraph="""At exactly six in the morning, the town of Kalluru woke up to a strange silence. The usual sounds of milk cans clinking, temple bells ringing, and buses coughing their way onto the main road were missing. Even the stray dogs, who normally announced every passing cyclist as if it were a national emergency, sat quietly as though they were waiting for something. Arjun noticed this first while standing on his balcony, holding a cup of tea that had already gone cold. The sky looked normal, the sun rose as it always did, yet the air felt unusually patient, as if time itself had slowed down.

Arjun worked as a signboard painter, a job that rarely surprised him anymore. Every shop wanted the same bold colours and the same predictable promises. That morning, however, when he unlocked his workshop, he found a wooden board leaning against the door. On it were words painted neatly in blue: *“Paint what you have never seen.”* There was no name, no address, only that sentence. Arjun frowned, because he was certain he had locked the shop properly the previous night.

Curiosity won over caution. He placed the board on his easel and dipped his brush into the paint. Without planning, his hand began to move. He painted a clock without numbers, a road that curved into the sky, and a tree whose leaves looked like small mirrors. As he worked, he felt a lightness he hadn’t felt in years, as if the brush were guiding him instead of the other way around. When he finally stepped back, the painting felt complete, though he could not explain why.

By afternoon, people began to gather outside the workshop. They stared at the board in silence, each person seeing something different. An old woman wiped her eyes, whispering that the tree looked like her childhood home. A schoolboy laughed and said the road was exactly how he imagined his future. Arjun realised that the painting wasn’t showing a single image but many, changing slightly with every pair of eyes that looked at it.

As evening arrived, the silence of the morning returned, gently this time. The wooden board was gone, as mysteriously as it had appeared, leaving behind only faint blue stains on the floor. Life in Kalluru returned to normal the next day, with buses honking and dogs barking again. Yet something had shifted. Arjun went back to painting ordinary signboards, but every now and then, he added a small, almost invisible detail — a curve, a colour, a hint of the impossible. And those who noticed it walked away smiling, feeling, for reasons they could not explain, that the world was slightly bigger than they had believed the day before.
"""
```

```python
documents = [paragraph]
vectorizer = CountVectorizer(stop_words="english")
X = vectorizer.fit_transform(documents)
df_bow = pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out()
)
print("Bag of Words DataFrame:")
print(df_bow)
```
