```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/User/Downloads/training_may/Machine Learning/datasets/netflix_titles.csv")
df.info()
```

```python
df.fillna({"director":"not specified",
           "cast":"not specified",
           "country":"not specified",
           "date_added":"not specified",
           "rating":"not specified",
           "duration":"not specified",
           },inplace=True)
```

```python
df.isnull().sum()
```

```python
df["date_added"]=pd.to_datetime(df["date_added"],errors="coerce")
df["year_added"]=df["date_added"].dt.year
df["month_added"]=df["date_added"].dt.month
```

```python
df.duplicated().sum()
```

```python
df.info()
df.isnull().sum()
```

```python
df["type"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

```python
df_time = df.dropna(subset=["year_added"])

df_time["year_added"].value_counts().sort_index()
```

```python
df_time["year_added"].value_counts().sort_index().plot(kind="line")
plt.title("Content Added to Netflix Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.show()
```

```python
df["country"].value_counts().head(10)
```

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Content-Producing Countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

```python
df["rating"].value_counts()
```

```python
df["rating"].value_counts().plot(kind="bar")
plt.title("Content Rating Distribution on Netflix")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns

sns.countplot(data=df, x="rating", hue="type")
plt.title("Rating Distribution by Content Type")
plt.xticks(rotation=45)
plt.show()


```
